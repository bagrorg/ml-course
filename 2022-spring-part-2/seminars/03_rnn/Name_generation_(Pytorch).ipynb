```python
all_letters = set()
for name in names:
  all_letters.update(name)

all_letters = ['START', 'STOP'] + list(all_letters)

n_letters = len(all_letters) 
n_letters
```

```python
char_to_int = dict([(b,a) for a,b in enumerate(all_letters)])
```

```python
def inputTensor(line):
    letter_indexes = [char_to_int[c] for c in ["START"] + list(line)]
    return torch.LongTensor(letter_indexes)

# LongTensor of second letter to end (EOS) for target
def targetTensor(line):
    letter_indexes = [char_to_int[c] for c in list(line) + ["STOP"]]
    return torch.LongTensor(letter_indexes)
```

```python
def random_training_example():
    line = names[random.randint(0, len(names) - 1)]
    input_line_tensor = inputTensor(line)
    target_line_tensor = targetTensor(line)
    return input_line_tensor, target_line_tensor
```

```python
class GeneratorNet(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(GeneratorNet, self).__init__()
        embeding_size = 32
        self.hidden_size = hidden_size

        self.embedding = nn.Embedding(input_size, embeding_size)
        self.rnn = nn.GRU(input_size=embeding_size, hidden_size=hidden_size, batch_first=True)
        self.output = nn.Linear(hidden_size, output_size)

    def forward(self, i, h0):
        emb = self.embedding(i)
        o, hn = self.rnn(emb, h0)
        o = self.output(o)
        o = nn.LogSoftmax(dim=-1)(o)
        return o, hn

    def initHidden(self):
        return torch.zeros(1, 1, self.hidden_size)
```

```python
criterion = nn.NLLLoss()


def train(optimizer, input_line_tensor, target_line_tensor):
    input_line_tensor = input_line_tensor.unsqueeze(0)
    target_line_tensor = target_line_tensor.unsqueeze(0)
   
    output, _ = rnn(input_line_tensor, rnn.initHidden())
    
    loss = criterion(output.flatten(0, 1), target_line_tensor.flatten())

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    return output, loss.item() / input_line_tensor.size(0)
```

```python
rnn = GeneratorNet(n_letters, 128, n_letters)

learning_rate = 0.001

optimizer = torch.optim.Adam(rnn.parameters(), lr=learning_rate)

n_iters = 100000
print_every = 5000
total_loss = 0 

for iter in range(1, n_iters + 1):
    output, loss = train(optimizer, *random_training_example())
    total_loss += loss

    if iter % print_every == 0:
        print('(%d %d%%) %.4f' % (iter, iter / n_iters * 100, total_loss / print_every))
        total_loss = 0
```

```python
def letterTensor(ch):
    return torch.LongTensor([[[char_to_int[ch]]]])
```

```python
max_length=30

def sample(start_letter='START'):
    with torch.no_grad():  
        input = letterTensor(start_letter)
        hidden = rnn.initHidden()

        output_name = ""

        for i in range(max_length):
            output, hidden = rnn(input[0], hidden)

            c = torch.distributions.categorical.Categorical(logits=output[0, 0])
            i = c.sample()

            if all_letters[i] == "STOP":
                break
            else:
              letter = all_letters[i]
              
              output_name += letter
            input = letterTensor(letter)

        return output_name
```
